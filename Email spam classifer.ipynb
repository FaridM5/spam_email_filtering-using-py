```python
emails=[folder+file for file in files]
```

```python
words=[]

for email in emails:
    f=open(email,encoding='latin-1')
    blob=f.read()
    words+=blob.split(" ")
```

```python
len(words)
```

```python
# To remove all words which are not used in our Analysis
for i in range(len(words)):
    if not words[i].isalpha():
        words[i]=""
```

```python
word_dict=Counter(words)
len(word_dict)
```

```python
del word_dict[""]
```

```python
word_dict=word_dict.most_common(3000)
```
